```python
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')
```

```python
a = pd.read_csv("/Users/pawelmiech/Downloads/roczne_wskazniki_cen_towarow_i_uslug_konsumpcyjnych_od_1950_roku.csv", encoding='utf8', usecols=[0,1], delimiter=";")
```

```python
a['inf'] = a.inflacja.apply(lambda x: x-100)
```

```python
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate("{:.2f}".format(round(height, 2)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
rects = ax.bar(a.Rok[-20:], a.inf[-20:])
ax.set_ylabel("Inflation (%)")
ax.set_xlabel("Year")
autolabel(rects)
plt.xticks(a.Rok[-20:])
plt.show()
```
